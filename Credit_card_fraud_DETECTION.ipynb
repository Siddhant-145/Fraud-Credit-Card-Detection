```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
print(credit_card_data.shape)
```

```python
# first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset informations
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
# separating the data for analysis
Non_fraud = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(Non_fraud.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
Non_fraud.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
Non_fraud_sample = Non_fraud.sample(n=492)
```

```python
new_dataset = pd.concat([Non_fraud_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
#confusion matrix for decision tree
conf_matrix = confusion_matrix(Y_test,X_test_prediction)
print(conf_matrix)
```

```python
#classification report for logistic regression
classification_rep = classification_report(Y_test, X_test_prediction)
print(classification_rep)
```

```python
model2 = DecisionTreeClassifier(random_state=42)
```

```python
model2.fit(X_train, Y_train)
```

```python
# accuracy on training data on decision tree
X_train_prediction_decision_tree = model2.predict(X_train)
training_data_accuracy_decision_tree = accuracy_score(X_train_prediction_decision_tree, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy_decision_tree)
```

```python
# accuracy on test data on decision tree
X_test_prediction_decision_tree = model2.predict(X_test)
test_data_accuracy_decision_tree = accuracy_score(X_test_prediction_decision_tree, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy_decision_tree)
```

```python
#confusion matrix for decision tree
conf_matrix_decision_tree = confusion_matrix(Y_test,X_test_prediction_decision_tree)
print(conf_matrix_decision_tree)
```

```python
#classification report for decision tree
classification_rep_decision_tree = classification_report(Y_test, X_test_prediction_decision_tree)
print(classification_rep_decision_tree)
```
